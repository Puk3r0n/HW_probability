##### Импорт необходимых библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

#### Пункт А

```python
B = 10000
outcomes = []

for _ in range(B):
  count = 0
  while True:
    count += 1
    roll = np.random.choice([3, 4, 5, 6, 7, 8])
    if roll == 8:
      outcomes.append(count)
      break

plt.hist(outcomes, bins=range(1, max(outcomes) + 2), density=True, edgecolor='black', alpha=0.75 )
plt.xlabel("Число бросков")
plt.ylabel("Частота")
plt.title("Гистограмма числа бросков кубика")
plt.show()
```

#### Пункт Б

```python
print(f"Количественная оценка математического ожидания числа бросков кубика {np.mean(outcomes)}")
```

#### Пункт В

```python

print(f"Количественная оценка вероятности окончания игры быстрее, чем за 5 бросков кубика ≈ {np.mean(np.array(outcomes) < 5)}")
```

#### Пункт Г

```python
B = 10000
outcomes_conditional = []

while len(outcomes_conditional) < B:
  count = 0
  valid_experiment = True
  while True:
    count += 1
    roll = np.random.choice(range(1, 9))
    if roll == 1 or roll == 2:
      valid_experiment = False
      break
    if roll == 8:
      if valid_experiment:
        outcomes_conditional.append(count)
      break

plt.hist(outcomes_conditional, bins=range(1, max(outcomes_conditional) +  1), density=True, edgecolor='black', alpha=0.75)
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.xlabel("Число бросков")
plt.ylabel("Частота")
plt.title("Условная гистограмма числа бросков октаэдра")
plt.show()
```

#### Пункт Д

```python
print(f"Оценка условного математического ожидания числа бросков октаэдра: {np.mean(outcomes_conditional)}")
```

#### Пункт Е

```python
print(f"Оценка условной вероятности окончания игры быстрее, чем за 5 бросков октаэдра (если грани 1 и 2 не выпадали): {np.mean(np.array(outcomes_conditional) < 5)}")
```
